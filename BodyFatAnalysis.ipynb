```python
##Calcilating BMI ----------> kg/m^2
W = []
H = []
for n in range(len(x.iloc[:,3])):
    m = x.iloc[n,3]
    W.append(m)
#print(W)
for n in range(len(x.iloc[:,4])):
    s = x.iloc[n,4]
    H.append(s)
#print(H)
W = np.array(W)
Weight = W/2.2
H = np.array(H)
Height = (.0254*H)*(.0254*H)
#print(Weight, Height)
bmi_calc = Weight/Height
print(bmi_calc)
```

```python
#New Column 
x["bmi"] = bmi_calc
```
